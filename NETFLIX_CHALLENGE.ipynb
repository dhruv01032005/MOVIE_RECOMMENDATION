```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import torch
import torch.optim as optim
from apyori import apriori
```

```python
data = pd.read_csv('/content/NETFLIX_MOVIE_RECOMMENDATION.csv', encoding='latin1', header=None)
```

```python
data
```

```python
transaction = (data.to_numpy().astype('str')).tolist()
```

```python
unique_movies = []
for i in transaction:
  for j in i:
    if j == 'nan':
      continue
    elif j not in unique_movies:
      unique_movies.append(j)
```

```python
np.random.seed(42)
dataset = []
for i in transaction:
  temp = [np.nan]*len(unique_movies)
  for k in i:
    if k == 'nan':
      continue
    else:
      temp[unique_movies.index(k)] = 5
  dataset.append(temp)

new_data = pd.DataFrame(dataset, columns = unique_movies)
```

```python
new_data
```

## Apriori

```python
intelligence = apriori(transactions = transaction, min_support=0.001, min_confidence=0.2, min_lift=2.5, min_length=2, max_length = 2)
```

```python
results = list(intelligence)
```

```python
movie_1 = []
movie_2 = []
support = []
confidence = []
lift = []
for result in results:
    movie_1.append(tuple(result[2][0][0])[0])
    movie_2.append(tuple(result[2][0][1])[0])
    support.append(result[1])
    confidence.append(result[2][0][2])
    lift.append(result[2][0][3])

combined = list(zip(movie_1, movie_2, support, confidence, lift))
```

```python
pairs = pd.DataFrame(combined, columns = ['Product 1', 'Product 2', 'Support', 'Confidence', 'Lift'])
pairs.sort_values('Lift', ascending = False, inplace = True)
pairs.reset_index(drop = True, inplace = True)
pairs
```

## Matrix Factorization

In this dataset I didn't know the peoples rating to each movie thus for this I have taken assumption that they have seen the movies that they like thus rated each movie as 5.

Thus this method in this case is biased thus not good method to predict someone's movie.

```python
# Modify the loss function to ignore NaN values

def factorize(A, k):

    # Hyperparameters
    epochs = 1000
    learning_rate = 0.01

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    A = A.to(device)

    # Randomly initialize A and B
    torch.random.manual_seed(42)
    W = torch.randn(A.shape[0], k, requires_grad=True, device=device)
    H = torch.randn(k, A.shape[1], requires_grad=True, device=device)

    # Optimizer
    optimizer = optim.Adam([W, H], lr=learning_rate)
    mask = ~torch.isnan(A)

    # Train the model
    for i in range(epochs):
        # Compute the loss
        #diff_matrix = 5*torch.nn.Sigmoid()(torch.mm(W, H)) - A
        diff_matrix = torch.mm(W, H) - A
        diff_vector = diff_matrix[mask]
        loss = torch.norm(diff_vector)

        # Zero the gradients
        optimizer.zero_grad()

        # Backpropagate
        loss.backward()

        # Update the parameters
        optimizer.step()

        if i % 100 == 0:
            print(f"Iteration {i}/{epochs}, Loss: {loss.item()}")

    return torch.mm(W, H)
```

```python
new_weights = factorize(torch.tensor(new_data.to_numpy()), 5)
```

```python
predicted_ratings = 5*torch.nn.Sigmoid()(new_weights).cpu().detach().numpy()
prediction  = pd.DataFrame(predicted_ratings, columns = unique_movies)
prediction
```

```python
def predict_movie_by_user(A, B, id):
    unique_movies = A.columns
    user_ratings = A.iloc[id]
    predicted_ratings = B.iloc[id]
    return pd.DataFrame({"Observed": user_ratings, "Predicted": predicted_ratings}, index=unique_movies)
```

```python
prediction_id = predict_movie_by_user(new_data, prediction, 5000)
prediction_id.sort_values('Predicted', ascending = False)
```
